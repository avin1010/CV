```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
I = cv2.imread('apple.jpg')
```

```python
I_rgb = cv2.cvtColor(I,cv2.COLOR_BGR2RGB)
```

```python
plt.subplot(2,2,1)
plt.imshow(I_rgb)
```

```python
I_c = cv2.bitwise_not(I_rgb)
```

```python
I_c
```

```python
plt.subplot(2,2,2)
plt.imshow(I_c)
```

```python
i_gray = cv2.cvtColor(I,cv2.COLOR_BGR2GRAY)
```

```python
plt.subplot(2,2,3)
plt.imshow(i_gray,cmap='gray')
```

```python
i_gray_c = cv2.bitwise_not(i_gray)
```

```python
plt.subplot(2,2,4)
plt.imshow(i_gray_c,cmap='gray')
```
