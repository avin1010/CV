```python
import cv2
import matplotlib.pyplot as plt
```

```python
I = cv2.imread('apple.jpg')
```

```python
I_rgb = cv2.cvtColor(I,cv2.COLOR_BGR2RGB)
```

```python
plt.subplot(3,2,1)
plt.imshow(I_rgb)
```

```python
#Red component

r = I_rgb[:,:,0]

plt.subplot(3,2,2)
plt.imshow(r,cmap='gray')
```

```python
# green component

g = I_rgb[:,:,1]

plt.subplot(3,2,3)
plt.imshow(g,cmap='gray')
```

```python
# blue component

b = I_rgb[:,:,2]

plt.subplot(3,2,4)
plt.imshow(b,cmap='gray')
```

```python
# gray scale

I_gray = cv2.cvtColor(I,cv2.COLOR_BGR2GRAY)

plt.subplot(3,2,5)
plt.imshow(I_gray,cmap='gray')
```
