```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('apple.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
width,height = image.shape[:2]
```

```python
bit_planes = np.zeros((8,width,height),dtype=int)
```

```python
for i in range(8):
    bit_planes[i] = (image>>i) & i
```

```python
plt.figure(figsize=(12, 6)) 
for i in range(8):
    plt.subplot(2, 4, i+1) 
    plt.imshow(bit_planes[i],cmap='gray')
    plt.title(f'Bit Plane {i+1}')
    plt.axis('off')
```
